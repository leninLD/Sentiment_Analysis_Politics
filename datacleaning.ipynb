# Data cleaning

```python
import re
import os
import json
import unicodedata
import pandas as pd
import numpy as np
from typing import List

class DataHandler:
    def __init__(self, root: str,
                 file_ext: str = "xlsx",
                 old_vocab: str = None,
                 min_count: int = 200,
                 max_word_len: int = 20, 
                 min_word_needed: int = 3):
        
        self.root         = root
        self.file_ext     = file_ext
        self.vocab        = {}
        self.old_vocab_path = old_vocab
        self.min_word_needed = min_word_needed
        self.filter_words = None
        self.min_count    = min_count
        self.max_word_len = max_word_len

        if old_vocab is not None:
            with open(old_vocab) as fp:
                self.vocab = json.load(fp)
                less_freq_words = [k for k, v in self.vocab.items() if v < min_count]
                huge_words      = [k for k in self.vocab.keys() if len(k) > max_word_len]
                self.filter_words = less_freq_words + huge_words
                self.vocab = {k: v for k, v in self.vocab.items() 
                              if k not in self.filter_words}

    # ── File Reader ───────────────────────────────────────────────────────────

    def read_files(self, number_of_files: int = 10):
        all_filenames = [
            f for f in os.listdir(self.root)
            if f.split(".")[-1] == self.file_ext
        ][:number_of_files]

        combined = pd.concat([
            pd.read_excel(os.path.join(self.root, f)) for f in all_filenames
        ])

        # Keep only text and label columns
        self.data = combined[['text', 'label']].copy()
        self.data = self.data.dropna(subset=['text', 'label'])
        self.data = self.data.reset_index(drop=True)

        print(f"Read {len(all_filenames)} files. Total rows: {len(self.data)}")
        print(f"Label Counts:\n{self.data['label'].value_counts()}")
        return self

    # ── Cleaning Helpers ──────────────────────────────────────────────────────

    def _rm_emojis(self, t: str) -> str:
        """Remove all emojis using unicode code point ranges + category fallback"""
        cleaned = []
        for char in t:
            cp = ord(char)
            if (
                0x1F600 <= cp <= 0x1F64F or
                0x1F300 <= cp <= 0x1F5FF or
                0x1F680 <= cp <= 0x1F6FF or
                0x1F700 <= cp <= 0x1F77F or
                0x1F780 <= cp <= 0x1F7FF or
                0x1F800 <= cp <= 0x1F8FF or
                0x1F900 <= cp <= 0x1F9FF or
                0x1FA00 <= cp <= 0x1FA6F or
                0x1FA70 <= cp <= 0x1FAFF or
                0x2600  <= cp <= 0x26FF  or
                0x2700  <= cp <= 0x27BF  or
                0x2300  <= cp <= 0x23FF  or
                0x2B50  <= cp <= 0x2B55  or
                0x25AA  <= cp <= 0x25FE  or
                0x1F1E0 <= cp <= 0x1F1FF or
                0x1F100 <= cp <= 0x1F1FF or
                0x1F200 <= cp <= 0x1F2FF or
                0x1F000 <= cp <= 0x1F02F or
                0x1F0A0 <= cp <= 0x1F0FF or
                0xFE00  <= cp <= 0xFE0F  or
                0x1F3FB <= cp <= 0x1F3FF or
                0x2702  <= cp <= 0x27B0  or
                cp == 0x200D             or
                cp == 0xFE0F             or
                cp == 0x00A9             or
                cp == 0x00AE             or
                cp == 0x2122             or
                cp == 0x3030             or
                cp == 0x303D             or
                cp == 0x3297             or
                cp == 0x3299             or
                unicodedata.category(char) in ('So', 'Sk')
            ):
                continue
            cleaned.append(char)
        return ''.join(cleaned)

    def _rm_zero_width(self, t: str) -> str:
        return re.sub(r'[\u200c-\u200f\u202a-\u202f\u2066-\u2069]', '', t)

    def _rm_urls(self, t: str) -> str:
    # Remove ALL urls aggressively
        t = re.sub(r'http[s]?://\S+', ' ', t)
        t = re.sub(r'www\.\S+', ' ', t)
        t = re.sub(r't\.co/\S+', ' ', t)           # ← Twitter short URLs
        t = re.sub(r'\b\S+\.co\.np\S*', ' ', t)    # ← Nepali domains
        t = re.sub(r'\b\S+\.(com|org|net|io|gov|edu|co|info|biz|np|me|tv)\S*',
                ' ', t, flags=re.IGNORECASE)
        return re.sub(r'\s+', ' ', t).strip()

    def _rm_mentions(self, t: str) -> str:
        t = re.sub(r'@\S*', '', t)
        return re.sub(r'\s+', ' ', t).strip()

    def _rm_hashtags(self, t: str) -> str:
        t = re.sub(r'#\S*', '', t)
        return re.sub(r'\s+', ' ', t).strip()

    def _rm_noise_chars(self, t: str, noise: str) -> str:
        t = t.lower()
        for char in noise:
            t = t.replace(char, ' ')
        return re.sub(r'\s+', ' ', t).strip()

    def _rm_special_chars(self, t: str) -> str:
        t = re.sub(r'[^\u0900-\u097Fa-zA-Z0-9\s]', ' ', t)
        return re.sub(r'\s+', ' ', t).strip()

    def _rm_stopwords(self, t: str, stop_words: List[str]) -> str:
        words = t.split()
        words = [w for w in words if len(w) > 1 and w not in stop_words]
        return ' '.join(words)

    # ── Core Preprocessor ─────────────────────────────────────────────────────

    def preprocess_tweet(self, tweet: str, noise: str, stop_words: List[str]) -> str:
        if tweet is None or str(tweet).strip() in ('', 'nan'):
            return ''

        t = str(tweet)
        t = self._rm_emojis(t)
        t = self._rm_zero_width(t)
        t = self._rm_urls(t)
        t = self._rm_mentions(t)
        t = self._rm_hashtags(t)
        t = self._rm_noise_chars(t, noise)
        t = self._rm_special_chars(t)
        t = self._rm_stopwords(t, stop_words)

        if self.old_vocab_path is None:
            for word in t.split():
                self.vocab[word] = self.vocab.get(word, 0) + 1

        return t.strip()

    # ── Main Cleaner ──────────────────────────────────────────────────────────

    def data_clean(self, stopwords_path: str, apply_filter: bool = False):
        noise = r'~؟॥"▬`%´•●=+÷।–][{}*"_…' + r"''\/" + ';@abcdefghijklmnopqrstuvwxyz1234567890०१२३४५६७८९()-.|!?",:—?।' + "'"

        with open(stopwords_path, encoding='utf-8') as fp:
            stop_words = [line.strip() for line in fp.readlines()]

        if apply_filter and self.filter_words and len(self.filter_words) > 0:
            stop_words += self.filter_words

        # Apply preprocessing
        self.data['clean_text'] = self.data['text'].apply(
            lambda x: self.preprocess_tweet(x, noise, stop_words)
        )

        # Drop duplicates and empty rows
        self.data = self.data.drop_duplicates(subset=['clean_text'])
        self.data = self.data[self.data['clean_text'].str.strip() != '']
        self.data = self.data.reset_index(drop=True)

        # Sort vocab
        self.vocab = dict(sorted(self.vocab.items(), key=lambda x: x[1]))

        # Apply filter words if needed
        if apply_filter:
            if not self.filter_words or len(self.filter_words) == 0:
                less_freq_words = [k for k, v in self.vocab.items() if v < self.min_count]
                huge_words      = [k for k in self.vocab.keys() if len(k) > self.max_word_len]
                self.filter_words = less_freq_words + huge_words

                self.data['clean_text'] = self.data['clean_text'].apply(
                    lambda x: self.filter_txt(x, self.filter_words)
                )
                self.vocab = {k: v for k, v in self.vocab.items()
                              if k not in self.filter_words}

        if self.filter_words:
            self.vocab = {k: v for k, v in self.vocab.items()
                          if k not in self.filter_words}

        # Remove rows with too few words
        self.data = self.data[
            self.data['clean_text'].apply(
                lambda x: len(x.strip().split()) >= self.min_word_needed
            )
        ]
        self.data = self.data.reset_index(drop=True)

        print(f"Cleaned data: {len(self.data)} rows | Vocab size: {len(self.vocab)}")
        print(f"Label Counts:\n{self.data['label'].value_counts()}")
        return self

    # ── Filter Text ───────────────────────────────────────────────────────────

    def filter_txt(self, text, filter_words):
        words = text.split()
        words = [w for w in words if w not in filter_words]
        return ' '.join(words)

    # ── Save Output ───────────────────────────────────────────────────────────

    def save(self, output_path: str):
        self.data[['text', 'label', 'clean_text']].to_csv(output_path, index=False)
        print(f"Saved {len(self.data)} rows to {output_path}")
        return self

    def __iter__(self):
        for _, row in self.data.iterrows():
            yield row['clean_text'].split(' ')
```
